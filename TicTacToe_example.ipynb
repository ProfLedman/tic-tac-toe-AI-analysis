```python
def stateOfBoard(board):
    # Possible winning lines in tic tac toe
    winning_lines = [
        (0,1,2), (3,4,5), (6,7,8),  # rows
        (0,3,6), (1,4,7), (2,5,8),  # columns
        (0,4,8), (2,4,6)            # diagonals
    ]

    # Check for a winner
    for a, b, c in winning_lines:
        if board[a] == board[b] == board[c] != 0:
            return board[a]  # 1 if O won, 2 if X won

    # Check if there are any empty spaces left
    if 0 in board:
        return -1  # The game is not finished yet

    # If no winner and no empty spaces, it's a draw
    return 0



def showBoard(board):
    S=[".","O","X"]
    print(S[board[0]],S[board[1]],S[board[2]])
    print(S[board[3]],S[board[4]],S[board[5]])
    print(S[board[6]],S[board[7]],S[board[8]])

def makeMove(board,location,player):
    board=list(board)
    board[location]=player
    return tuple(board)

startBoard=(0,0,0,0,0,0,0,0,0)

ourBoard=(0,0,0,0,0,0,0,0,0)
ourBoard=makeMove(ourBoard,4,1)
print(ourBoard)
showBoard(ourBoard)
print(ourBoard.count(1))

print("board state",stateOfBoard((1,2,1,0,2,0,1,0,0)))
print("board state",stateOfBoard((1,1,1,0,0,0,2,2,2)))
print("board state",stateOfBoard((2,2,2,0,0,0,0,0,0)))
```

```python
def getEmpty(board):
    empty=[]
    for i in range(9):
        if board[i]==0:
            empty.append(i)
    return empty

def play(policyA,policyB,visualize=False):
    board=(0,0,0,0,0,0,0,0,0)
    nextPlayer=[0,2,1]
    player=1
    history=list()
    while stateOfBoard(board)==-1:
        #generate move
        locations=getEmpty(board)
        if player == 1:
            chosenLocation=numpy.random.choice(locations,p=policyA[board])
        else:
            chosenLocation=numpy.random.choice(locations,p=policyB[board])
        history.append([board,locations.index(chosenLocation)]) #remember!!!!
        #apply move
        board=makeMove(board,chosenLocation,player)
        #toggle player
        player=nextPlayer[player]
        if visualize:
            showBoard(board)
            print("----")
    return stateOfBoard(board),history

outcome,history=play(randomPolicy,randomPolicy)
print("outcome",outcome)
print("history",history)
```
